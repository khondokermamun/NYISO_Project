```python
import pandas as pd
from pathlib import Path

# 1. Define the root directory (using the raw string format to handle Windows paths)
base_dir = Path(r'E:\NYISO_Project\dataset')

# 2. Find all CSV files recursively in all subfolders
csv_files = list(base_dir.rglob('*.csv'))
print(f"Found {len(csv_files)} CSV files. Reading data...")

# 3. Read each CSV into a list of DataFrames
df_list = []
for file in csv_files:
    try:
        # Read the CSV
        df = pd.read_csv(file)
        df_list.append(df)
    except Exception as e:
        print(f"Failed to read {file.name}: {e}")

# 4. Concatenate all daily/monthly DataFrames into one massive DataFrame
print("Concatenating files into a single timeline...")
full_df = pd.concat(df_list, ignore_index=True)

# 5. Display the first few rows and column names to verify
print("\nData Combination Complete!")
print(f"Total rows: {len(full_df)}")
print(f"Columns found: {full_df.columns.tolist()}")

# Display the first 5 rows
print(full_df.head())
```

```python
# 6. Save the combined DataFrame to a new CSV file
full_df.to_csv(base_dir / 'combined_data.csv', index=False)
```
